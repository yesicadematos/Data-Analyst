```python
import sqlalchemy as sa
import pandas as pd 
from sqlalchemy import text
from conector import enginok
```

**script de cosulta para ver el resultado total de Pantallas,Butacas ,Espacio_Incaa por "Provincia"** 

```python
text=("""SELECT "Provincia" , sum ("Pantallas")as "Pantallas" ,sum("Butacas")as "Butacas" ,count("Espacio_Incaa") as "Espacio_Incaa" 
FROM "CINES"
NATURAL JOIN "Departamento"
NATURAL JOIN "Localidad"
NATURAL JOIN "Provincia"
GROUP BY "Provincia"
ORDER BY "Provincia"   ASC;""")
df_query = pd.read_sql(text, con=enginok)
```

**creacion de VIEW para hacer otras consultas sobre ellas**

```python
text=("""CREATE OR REPLACE VIEW public.total_registro
AS
SELECT "CINES"."Categoria",
"CINES"."Fuente"
FROM "CINES";""")
enginok.execute(text) 
```

```python
text=("""CREATE OR REPLACE VIEW public.total_registro2
AS
SELECT "MUSEOS"."Categoria",
"MUSEOS"."Fuente"
FROM "MUSEOS";""")
enginok.execute(text)  
```
