```python
import sqlalchemy as sa
import pandas as pd 
from sqlalchemy import text
from conector import enginesexy
```

```python
text=("""SELECT sum ("Pantallas")as "Pantallas" ,sum("Butacas")as "Butacas" ,count("Espacio_Incaa") as "Espacio_Incaa" , "Provincia"
FROM "CINES"
NATURAL JOIN "Departamento"
NATURAL JOIN "Localidad"
NATURAL JOIN "Provincia"
GROUP BY "Provincia"
ORDER BY "Provincia"   ASC;""")
df_query = pd.read_sql(text, con=enginesexy)
```

```python
text=("""CREATE OR REPLACE VIEW public.total_registro
AS
SELECT "CINES"."Categoria",
"CINES"."Fuente"
FROM "CINES";""")
enginesexy.execute(text) 
```

```python
text=("""CREATE OR REPLACE VIEW public.total_registro2
AS
SELECT "MUSEOS"."Categoria",
"MUSEOS"."Fuente"
FROM "MUSEOS";""")
enginesexy.execute(text)  
```

```python
text=("""CREATE OR REPLACE VIEW public.total_registro3
AS
SELECT "BIBLIOTECA"."Categoria",
"BIBLIOTECA"."Fuente"
FROM "BIBLIOTECA";""")
enginesexy.execute(text)  
```
