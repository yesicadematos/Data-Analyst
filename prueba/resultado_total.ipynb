```python
df_total_parcial=df_totalB.append(df_totalC ,sort=False,ignore_index=False)
# .append(df_totalM  , ignore_index=False)
```

```python
df_total = df_total_parcial.append(df_totalM, sort=False,ignore_index=False)
df_total.tail()
```

con los datos en conjunto podemos ver el registro total por 'Categoria' y 'Fuente' utilizando Groupby

```python
df_total.groupby(['Fuente','Categoria'])['Id'].count()
```

con el mismo conjunto de datos podemos ver los registros totales pero esta ves por 'Provincia' y 'Categoria', pero ahora utilizamos .Concat y optenemos los mismos resultados.

```python
df_resultado=pd.concat([df_totalC,df_totalB,df_totalM], sort=False, ignore_index=False)
df_resultado.sample(6)
```

```python
df_resultado.groupby(['IdProvincia','Categoria'])[['Id']].count()
```
